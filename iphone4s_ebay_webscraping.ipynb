```python
    import pandas as pd
    from bs4 import BeautifulSoup as bs
    import requests

    headers={'user-agent':'please add your user agent'}
#     url="https://www.ebay.com/urw/Apple-iPhone-4s-16GB-Smartphone-Black-Unlocked-/product-reviews/109303245?pgn="
```

```python

page = 0
titles = []
title2 = []
while page != 75:
      url = f"https://www.ebay.com/urw/Apple-iPhone-4s-16GB-Smartphone-Black-Unlocked-/product-reviews/109303245?pgn={page}"
      response = requests.get(url)
      html = response.content
      soup = bs(html, "lxml") 
      for h3 in soup.find_all("p", class_="review-item-content rvw-wrap-spaces"):
#             import pdb; pdb.set_trace()
            titles.append(h3.get_text(strip=True))
    
      for h3 in soup.find_all("span", class_="show-full-review"):
            title2.append(h3.get_text(strip=True))
      page = page + 1
titles
```

```python
review1=pd.DataFrame(titles,columns=['Review1'])
review1
```

```python
r1dic = review1.to_dict()
r1dic
```
